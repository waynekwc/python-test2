```python
from finlab.data import Data
import pandas as pd
from finlab import ml
import talib
import time
import numpy as np

data = Data()

rev = data.get("當月營收")
rev.index = rev.index.shift(1, "d")  #晚五天才知道營收

revenue_yoy = round(100*((rev/rev.shift(12))-1),2)
revenue_yoy_average3 = revenue_yoy.rolling(3).mean()
revenue_yoy_average12 = revenue_yoy.rolling(12).mean()

T3395營業利益 = data.get('營業利益（損失）', 1000).fillna(data.get('營業利益', 1000))
T3100營業收入淨額 = data.get("營業收入合計", 1000)
R106_營業利益率 = T3395營業利益 / T3100營業收入淨額 *100

T3295營業毛利 = data.get('營業毛利（毛損）', 1000).fillna(data.get('營業毛利（毛損）淨額', 1000))
R105_營業毛利率 = T3295營業毛利 / T3100營業收入淨額 *100

grossratio_yoy = round(100*((R105_營業毛利率/R105_營業毛利率.shift(4))-1),2)
profitratio_yoy = round(100*((R106_營業利益率/R106_營業利益率.shift(4))-1),2)

T0100流動資產 = data.get("流動資產合計", 1000)
T0960非流動資產 = data.get("非流動資產合計", 1000).fillna(data.get("非流動資產淨額", 1000))
T0010資產總額 = T0100流動資產 + T0960非流動資產
T1000負債總額 = data.get("負債總額", 1000).fillna(data.get("負債總計", 1000))
R505_負債比率 = T1000負債總額 / T0010資產總額 * 100

debt_smaller_than_50_oneyear = (R505_負債比率<50) & (R505_負債比率<50).shift(1) & (R505_負債比率<50).shift(2)

T3970經常稅後淨利 = data.get('本期淨利（淨損）',1000).fillna(data.get('本期稅後淨利（淨損）',1000))
T3100營業收入淨額 = data.get("營業收入合計", 1000)
R108_稅後淨利率 = T3970經常稅後淨利 / T3100營業收入淨額 *100

T2000權益總計 = data.get('權益總計', 1000).fillna(data.get('權益總額', 1000))
T3971本期綜合損益總額 = data.get("本期綜合損益總額", 1000).fillna(data.get("綜合損益總額", 1000))
R11U_ROE綜合損益 =  (T3971本期綜合損益總額/((T2000權益總計+T2000權益總計.shift())/2)) * 100

股本 = data.get('股本合計', 1000)
T7210營運現金流 = data.get("營運產生之現金流入（流出）", 1000)
R303_每股現金流量 = T7210營運現金流 / 股本 * 10
R303_每股現金流量_average4 = R303_每股現金流量.rolling(4).mean()

close = data.get("收盤價")
adj_close = data.get_adj('收盤價')

close = close.fillna(method='backfill')
adj_close = adj_close.fillna(method='backfill')

para = 7

recent_rev = (rev.rolling(para).mean()>rev.rolling(para).mean().shift(1))

close_20 = close.rolling(20).mean()
close_60 = close.rolling(60).mean()
close_120 = close.rolling(120).mean()
close_240 = close.rolling(240).mean()

features = {
    'close': close,
    'close_20': close_20,
    'close_60': close_60,
    'close_120': close_120,
    'close_240': close_240,
    'adj_close': adj_close,
    'grossratio_yoy': grossratio_yoy,
    'profitratio_yoy': profitratio_yoy,
    'revenue_yoy': revenue_yoy,
    'revenue_yoy_average3': revenue_yoy_average3,
    'revenue_yoy_average12': revenue_yoy_average12,
    'recent_rev': recent_rev,
    'recent_rev2': recent_rev & recent_rev.shift(1),
    'recent_rev3': recent_rev & recent_rev.shift(1) & recent_rev.shift(2),
    'debt_smaller_than_50_oneyear': debt_smaller_than_50_oneyear,
    'R108_稅後淨利率': R108_稅後淨利率,
    'R11U_ROE綜合損益': R11U_ROE綜合損益,
    'R303_每股現金流量_average4': R303_每股現金流量_average4,
}

trading_frequency = rev.index

for name, f in features.items():
    features[name] = f.reindex(trading_frequency, method='ffill')
    
for name, f in features.items():
    features[name] = f.unstack()
    
dataset = pd.DataFrame(features)
dataset = dataset.reset_index().set_index("date")
# dataset.loc[:,'close':'adj_close'] = dataset.loc[:,'close':'adj_close'].fillna(method='backfill')
```

```python
from finlab.data import Data
import pandas as pd
from finlab import ml


data = Data()

rev = data.get("當月營收")
rev.index = rev.index.shift(1, "d")  #晚五天才知道營收
close = data.get("收盤價")
# close_per_week = close[::5]
adj_close = data.get_adj('收盤價')

close = close.T[(close.iloc[-150:].T.mean(axis=1)>70)].T
temp = close[::5]

# testlist = ['2454','2330']
# testclose = close[testlist]


# ===== trading frequency =====

every_month = rev.index
every_week = temp.index
trading_frequency = every_week

# ===== trading frequency =====

# ===== 計算features =====

def bias(n): #乖離率指標
    return close / close.rolling(n, min_periods=1).mean()

def acc(n): #加速度指標
    return close.shift(n) / (close.shift(2*n) + close) * 2

def rsv(n): #RSV指標
    l = close.rolling(n, min_periods=1).min()
    h = close.rolling(n, min_periods=1).max()
    
    return (close - l) / (h - l)

def mom(n):
    return (rev / rev.shift(1)).shift(n)

def yoy(n):
    return (rev / rev.shift(12)).shift(n)

def yoyacc(n):
    return ( (rev / rev.shift(12))/(rev / rev.shift(12)).shift(1) ).shift(n)

features = {
    'mom1': mom(1),
    'mom2': mom(2),
    'mom3': mom(3),
    'mom4': mom(4),
    'mom5': mom(5),
    'mom6': mom(6),
    'mom7': mom(7),
    'mom8': mom(8),
    'mom9': mom(9),  #why只到 mom 9???
    
    'bias5': bias(5),
    'bias10': bias(10),
    'bias20': bias(20),
    'bias60': bias(60),
    'bias120': bias(120),
    'bias240': bias(240),
    
    'acc5': acc(5),
    'acc10': acc(10),
    'acc20': acc(20),
    'acc60': acc(60),
    'acc120': acc(120),
    'acc240': acc(240),
    
    'yoy1': yoy(1),
    'yoy2': yoy(2),
    'yoy3': yoy(3),
    'yoy4': yoy(4),
    
    'yoyacc1': yoyacc(1),
    'yoyacc2': yoyacc(2),
}

# ===== 計算features =====

# ===== 組dataset =====

for name, f in features.items():
    features[name] = f.reindex(trading_frequency, method='ffill')
    
for name, f in features.items():
    features[name] = f.unstack()
    
dataset = pd.DataFrame(features)

feature_names = list(dataset.columns)

ml.add_profit_prediction(dataset)
ml.add_rank_prediction(dataset)

#神經網路容易被極值影響的網路, 此範例中刪除前百分之一

dataset_drop_extreme_case = ml.drop_extreme_case(dataset, 
    ['bias60', 'bias120', 'bias240', 'mom1', 'mom2', 'mom3', 'mom4', 'mom5', 'mom6'], thresh=0.01)

dataset_dropna = dataset_drop_extreme_case.dropna(how='any')
dataset_dropna = dataset_dropna.reset_index().set_index("date")

dataset_train = dataset_dropna[:'2015']
dataset_test = dataset_dropna['2016':]


```

```python
from finlab.data import Data
import datetime
import talib
import pandas as pd

#抓取股票代號
stockID='0050'

# 可以從 data 中讀取資料
data = Data()

period = 600

# 從 close, high, low 取從最新的一筆資料往前算，2000筆資料
close = data.get('收盤價', period)
high = data.get('最高價', period)
low = data.get('最低價', period)

print("抓取資料完成！")
```

```python
# 利用 talib 來計算這檔股票的指標
kd = talib.STOCH(high[stockID].ffill().values, 
                 low[stockID].ffill().values, 
                 close[stockID].ffill().values,
                 fastk_period=9, slowk_period=3
                 , slowd_period=3, slowk_matype=1, slowd_matype=1)

# 將 kd 值 分別轉換成 k、d 兩條 series
k = pd.Series(kd[0], index=close[stockID].index)
d = pd.Series(kd[1], index=close[stockID].index)

#買進訊號為KD黃金交叉，且K小於30
buy = (k > d) & (k.shift(1) < d.shift(1)) & (k < 30)
#賣出訊號為KD死亡交叉，且K大於70
sell = (k < d) & (k.shift(1) > d.shift(1)) & (k > 70)

#secondary_y代表拉長畫線到y軸，讓線更明顯
buy.astype(int).plot(secondary_y=True, color='red')
sell.astype(int).plot(secondary_y=True, color='green')
#畫出股價
close[stockID].plot()

#持有訊號
hold = pd.Series([0]*len(buy), index=buy.index)
hold[buy]=1
hold[sell]=-1
hold.plot()
```
